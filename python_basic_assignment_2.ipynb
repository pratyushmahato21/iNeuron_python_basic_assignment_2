2. What are the three different types of Boolean operators?

Ans: The three different types of Boolean operators are AND, OR, NOT.
Suppose we have two different conditions. Now if both conditions are True then it is AND statement. If only one statement is True then it is OR statement.
Now suppose we have only one condition, then there will only two scenarios, either the condition is satsfied of is not satisfied. This is a NOT statement.

8. Write code that prints Hello if 1 is stored in spam, prints Howdy if 2 is stored in spam, and prints Greetings! if anything else is stored in spam.

```python
spam = int(input("Enter a number: "))

if spam == 1:
    print("Hello")
    
elif spam == 2:
    print("Howdy")
    
else:
    print("Greetings!")
```
